```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib import markers
from math import sqrt
```

```python
def attract(x,t,u, parameters={}):
    dx = np.zeros((2,))
    dx[0] = x[1]+0.1*u
    dx[1] =-x[0]+1/3*x[0]**3-x[1]+0.1*u
    return dx
```

```python
initial=np.array([1,1])      
final=np.array([0,0])
```

```python
K = np.matrix([0.1,0.1])
u0 = np.matrix([0,0])
def controller(x, ref):
    cin = (u0-K @ (x - ref))[0,0]
    return cin
```

```python
#lqr
# Time interval
t = np.linspace(0,10,500)
# Store results for plotting 
u = np.zeros(len(t))

# control limit
op_hi = 100
op_lo = -100

x0 = initial
res1 =np.ones(len(t))
res2= np.ones(len(t))
res1[0] = initial[0]
res2[0] = initial[1]
```

```python
for i in range(len(t)-1):
    ts = [t[i],t[i+1]]
    u[i+1] = controller(x0, final)
    if u[i+1]>op_hi:
        u[i+1] = op_hi
    elif u[i+1]<op_lo:
        u[i+1] = op_lo
    y = odeint(attract, x0, ts, args=(u[i+1],))
    res1[i+1]=y[-1][0]
    res2[i+1]=y[-1][1]
    x0 = y[-1]
    print(x0)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# ax.set_ylim(79,83)
# ax.set_xlim(7.8,10)
ax.plot(t,res1)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
# ax.set_ylim(79,83)
# ax.set_xlim(7.8,10)
ax.plot(t,res2)
```
